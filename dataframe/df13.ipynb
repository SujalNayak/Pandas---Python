```python
import pandas as pd
```

```python
people1 = {"First Name": ["Sujal", "Rahul", "Raj", "Samar"], "Last Name": ["Shah", "Sharma", "Singh", "Sinha"], "Email": ["sujal.shah@example.com", "rahul.sharma@example.com", "raj.singh@example.com", "samar.sinha@example.com"]}
people1
```

```python
df = pd.DataFrame(people1)
df
```

```python
df["Full Name"] = df["First Name"] + " " + df["Last Name"]
df
```

```python
df.drop(columns=["First Name","Last Name"],inplace=True)
df
```

```python
df[["First","Last"]] = df["Full Name"].str.split(" ",expand=True)
df
```
