```python
import pandas as pd
```

```python
dfx = pd.DataFrame({"Fruit": ["Apple", "Banana", "Cherry", "Orange"], "Vegitables": ["Carrot", "Potato", "Tomato", "Cucumber"], "MarketPrice": [100, 50, 150, 80]})
dfx
```

```python
dfy = pd.DataFrame({"Fruit": ["Apple", "Banana", "Cherry", "Grapes"], "Vegitables": ["Carrot", "Potato", "Tomato", "Spinach"], "ShopPrice": [100, 50, 150, 120]})
dfy
```

```python
x = pd.merge(dfx, dfy, on=["Fruit","Vegitables"], how="inner")
x
```

```python
x = pd.merge(dfx, dfy, on=["Fruit","Vegitables"], how="outer")
x
```
