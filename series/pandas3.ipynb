```python
import pandas as pd
```

```python

s = pd.Series([10,20,30,40,50],index=['a','b','c','d','e'])
print(s)
```

```python
print(s['a'])
```

```python
print(s[0])
```

```python

#loc
print(s.loc['c'])
```

```python
#iloc
print(s.iloc[2])
```

```python

print("at",s.at['c'])
```

```python
print("iat",s.iat[2])
```

```python

#label based
print(s[['a','c']])
```

```python
print(s['a':'c'])
```

```python

#boolean access
print(s[s>30])
```
