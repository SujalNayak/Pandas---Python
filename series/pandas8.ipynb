```python
import pandas as pd
```

```python

s = pd.Series([10,20,30,0,None,40,None,50])
print(s)
```

```python
print(s.isnull())
```

```python
print(s.isnull().sum())
```

```python
print(s.notnull())
```

```python
print(s.notnull().sum())
```

```python

s = s.fillna(0)
print(s)
```

```python
print(s.isnull().sum())
```

```python

s = pd.Series([10,20,30,0,None,40,None,50])
print(s)
```

```python
s = s.dropna()
print(s)
```
